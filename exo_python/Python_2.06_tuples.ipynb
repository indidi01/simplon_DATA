### Immuable ou not immuable, tel est la question

![](https://miro.medium.com/max/1316/1*uFlTNY4W3czywyU18zxl8w.png)

Une variable (ou objet) est immuable si on ne peut pas le modifier après sa création. Il faut recréer un autre objet pour cela, ou stocker la modification dans une nouvelle variable. On appelle parfois ces variables des **constantes**.

Tu ne l'as peut-être pas remarqué, mais contrairement aux variables immuables (voir celles qui sont cochées dans l'image ci-dessus) les listes et les dictionnaires sont mutables. C'est à dire qu'on peut remplacer un élément par un autre.

```python
>>> liste = ['a', 'b', 'c']
>>> liste[1] # On accède au deuxième élément de la liste
'b'
>>> liste[2] # Troisième élément
'c'
>>> liste[1] = 'B' # On remplace 'b' par 'B'
>>> liste
['a', 'B', 'c']

```

On ne peut faire ça que parce que les variables de type **list** sont mutables.

Il est parfois utile d'avoir des variables immuables. Par exemple, des ingénieurs aéronautiques qui travaillent sur le programme d'un nouveau moteur d'avion ne voudraient pas que la valeur de [Pi](http://recordsdumonde.fr/record-de-memorisation-du-nombre-pi/) soit modifiée par erreur durant l'exécution. L'immutabilité garantie qu'elle ne peut pas l'être.

Un objet immuable permet parfois d'améliorer les performances d'un programme, ce qui est intéressant quand on doit effectuer un calcul sur une grande quantité de données.

On peut donc dire que l'immutabilité est un pattern... ou un anti pattern, en fonction de ton obédience.


### Les tuples

Les tuples ne sont rien d'autres que le pendant **immuable** des listes qui elles sont mutables. C'est donc une structure de données permettant de regrouper des données dans une boîte.

Elles permettent d'effectuer certaines choses comme sur une liste. On peut les parcourir, accéder à un élément, effectuer des calculs (par exemple additionner tous les nombres)...

![](https://d2h0cx97tjks2p.cloudfront.net/blogs/wp-content/uploads/sites/2/2017/12/Python-Tuple-Functions-01.jpg)

Un dictionnaire s'écrit avec des accolades {dictionnaire}, une liste avec des crochets [liste], un tuple avec des parenthèses (tuple). La notion de dimension fonctionne aussi avec les tuples.

Une dernière chose. Je t'ai menti, ou plutôt j'ai pas été complètement exact. Les tuples en Python ne sont pas complètement immuables. Pour en savoir plus voir les ressources.
