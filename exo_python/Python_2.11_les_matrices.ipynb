### Introduction

Dans cette quête, tu vas apprendre à faire des calculs matriciels. Il est important de bien comprendre cette notion de matrice, avant de commencer à utiliser la bibliothèque NumPy. Tu découvriras dans les quêtes suivantes en quoi le calcul matriciel à une importance fondamentale pour l'analyse de données.
