L'image ci-dessus montre la différence essentielle qu'il y a entre les dictionnaires et les listes. Quand dans le cas de la liste on accède à un élément grâce à son index, dans le cas d'un dictionnaire on y accède grâce à sa clé. Le dictionnaire fonctionne donc comme un système **clé-valeur**.

" Pourquoi utiliser une clé plutôt qu'une valeur numérique ? J'aimais bien cette option moi", me diras tu. Cela dépend de la situation. Je vais te donner un exemple pour illustrer l'intérêt du dictionnaire.

Imagine que tu veux ranger des données alimentaire par groupes, par exemple des catégories d'aliments (légumes, viandes, féculents...). Les ranger par numéro n'est pas le plus pratique, parce que tu vas avoir besoin d'un document faisant la correspondance entre les numéros pour te rappeler que 1 c'est pour les légumes, 2 pour les viandes, etc. Avec un dictionnaire tu indiqueras directement "viandes" et python te renverra tous les aliments de la catégorie demandée.

```python
food_fridge_dict = {"viandes": ["boeuf", "lapin", "agneau"],
             "légumes": ["courgette", "aubergine", "oignon"]}

food_fridge_list = [["boeuf", "lapin", "agneau"], ["courgette", "aubergine", "oignon"]]

```

Quels sont les viandes disponible dans mon frigo ?

```python
# simplement
>>> food_fridge_dict["viandes"]
["boeuf", "lapin", "agneau"]

# fastidieusement (encore plus quand la liste est longue)
>>> food_fridge_list[0]
["boeuf", "lapin", "agneau"]

```

Bref, comme tu l'as compris avec ce petit exemple, python te permet de faire différentes choses de différentes façons.

Tout comme une liste, un dictionnaire est une boîte. Il peut donc contenir tout ce qu'on veut. Il peut même avoir d'autres dictionnaires comme valeur ! Mais la clé sera toujours alphanumérique.

```python
house_and_food_dict = {"nourriture": {"légumes": ["courgette", "aubergine", "oignon"],
                                      "viandes": ["boeuf", "lapin", "agneau"]},
                       "mobilier": {"chambre": ["lit", "bureau", "chaise"],
                                    "salon": ["télévision", "table", "bar"]}}
# Comment accéder à la liste de légumes ?
>>> house_and_food_dict["nourriture"]["légumes"]
["courgette", "aubergine", "oignon"]
# Comment accéder au mobilier de salon ?
>>> house_and_food_dict["mobilier"]["salon"]
["télévision", "table", "bar"]

```
