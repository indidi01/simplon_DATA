### Ex 1.3 — Boucles et list comprehension
A partir de `prix_list = [45000, 180000, 350000, 620000, 1200000, 2500000]` :
1. Affiche chaque prix avec un `for`
2. Cree `prix_eleves` avec les prix > 500 000 (list comprehension)
3. Calcule la moyenne de `prix_list` sans numpy

```python
# TON CODE ICI
prix_list = [45000, 180000, 350000, 620000, 1200000, 2500000]
prix_eleves = []

for prix in prix_list:
    print(prix)
    if prix > 500000:
        prix_eleves.append(prix)
print(prix_eleves)

moyenne = sum(prix_list) / len(prix_list)
print("la moyenne =", moyenne)
```

### Ex 1.4 — Fonctions
Ecris `categoriser_bien(surface, nb_pieces)` qui retourne :
- 'Studio' si nb_pieces == 1
- 'Petit' si surface < 40
- 'Moyen' si surface < 80
- 'Grand' si surface < 120
- 'Tres grand' sinon

Teste avec au moins 5 cas.

```python
# TON CODE ICI
def categoriser_bien(surface, nb_pieces):
    if nb_pieces == 1:
        return"studio"
    if surface < 40:
        return"petit"
    elif surface < 80:
        return"moyen"
    elif surface < 120:
        return"grand"
    else:
        return"tres grand"

print(categoriser_bien(79, 1))
print(categoriser_bien(79, 2))
print(categoriser_bien(135, 4))
print(categoriser_bien(38, 5))
print(categoriser_bien(56, 6))
```
